```python
# I want to print the top ten values, not all together, but one by one.

class TopTen:
    
    def __init__(self): # Specifying a counter/counter variable here: num
        self.num = 1 # Wanna start from 1.
        
# When you say you want to create your own iterator you need two important methods the first one is iter method now it will
# give you the object of iterator the next method you need is next which will give you the next value or next object.

    def __iter__(self):
        return self
    
    def __next__(self): # We want the next value of num
        if self.num <= 10:
            val = self.num
            self.num += 1
            return val
        else:
            raise StopIteration
    
values = TopTen()

print(values.__next__())
```

```python
# Generators:
# We have talked about an iterator so normally we use iterator to print or to fetch one value at a time so let's say if you have
# a list of value say 100 values and if you want one by one value what makes sense is to use an iterator but there are small
# issues with an iterator right the first one you need to get the iterator where you have to define those two functions next and
# iter we don't want to do that we want Python to do it for us and that's where Python says ok if you want it iterators we will
# give you generators ok but now you will think we have two different concepts and not exactly generators will give you 
# iterators our job is done right so generator gives you an iterator but how exactly to create a generator so let's do that:
```

```python
# Define a function not a class this time and we'll do the same thing - return the top 10 values. topten() is the function 
# which will give you the iterator. 


# Normal function uses return:

def topten():
    return 5
    
values = topten()
print(values)
```

```python
# Generator: It gives an iterator - it isn't a normal function because we are using yield a. Yield also returns a value but in
# the format of an iterator. If you want to fetch something from an iterator we have to use the next function.

def topten():
    yield 5 # yield keyword makes our function, a generator.
    
values = topten()
print(values) # instead of getting 5 we got an object of generator
```

```python
# Even return gives the same output. Diff: Since it's yeild, it'll give you an iterator - you can write multiple yield statements.

def topten():
    yield 1 
    yield 2 
    yield 3 
    yield 4 
    
values = topten()

print(values.__next__()) # or can use print(next(values))
```

```python
# Eg, I wanna print the top ten perfect squares:

def TopTen():
    n = 1
    while n <= 10: # Not taking for loop but can take. Not taking for as indirectly for uses an iterator.
        sq = n * n
        yield sq # We are sending the value - so it is almost same as return but return will terminate the function and yeild won't.
        n += 1 # to iterate n
        
values = TopTen()

for i in values: 
    print(i)
```

```python
class Hello:
    def run(self):
        for i in range(5):
            print("Hello")

class Hi:
    def run(self):
        for i in range(5):
            print("Hi")
            
# To execute these methods we create objects of it:

t1 = Hello()
t2 = Hi()

t1.run()
t2.run()
```

```python
from threading import * # To use threads

class Hello(Thread): # One thread
    def run(self):
        for i in range(5):
            print("Hello")

class Hi(Thread): # One thread
    def run(self):
        for i in range(5):
            print("Hi")

t1 = Hello()
t2 = Hi()

t1.run()
t2.run()

# before running this what we are expecting is so you can see the diagram (picture) here so we have a main thread by default and
# main thread will execute all the statements the moment you say run (t1.run()) and run (t2.run()) here it should create two
# different threads which is let's say t1 and t2 and t1 will print hello five times and t2 will print hi five times and it 
# should be happening simultaneously so we should get the output as hello hi hello hi but let's run this code and let's see what
# happens:
```

```python
# Bubble Sort: In almost all sorting techniques, swapping is common - that is how we make sure all the values in our list is sorted.

# After doing the first iteration - we get the biggest value at the end - while sorting in ascending order. - now we reiterate
# the same thing - see pic. After 2nd iteration the second last value is the second big element.

# One thing which is important for the second iteration we are not checking for 8 right because it is done so the maximum value
# you have is 8 at the end right and after 2nd iteration the second last value is the second big element if you can compare the
# list now 7 and 8 those are the maximum values so after second iteration you got two big values in ascending order right that
#  means after all the iterations so number of elements we have here is 6 right so after fifth iteration you will get all the 
# values sorted right if you repeat the same steps that's what bubble sort is and that's why we require two loops one for the 
# iteration which will make sure that you will get the biggest element at the end and the second loop which is the outer loop
# will be for doing the same thing repeatedly and that's what you have to do in this code:
```

```python
# We can define sort method in some other module or here itself:

def sort(nums):
    # at the end it is all about swapping so we take 2 loops:
    for i in range(len(nums)-1,0,-1): # Doing it in descending order (iteration = -1 here) - can do it in ascending order too.
        # Going from 5 to 1
        for j in range(i): # so for the first iteration if we remember we have to go till end right you have to start with 0 and
# you have to reach till 5 in the next iteration you have to go from zero to four in the next iteration we have to go from zero
# to three after every internal iteration you will get the max value at the end then second last value will be the second
# biggest value third last value will be the third biggest value and that's why the range will go to i and not five it will go
# till i so that depends upon the value of I so it is 5 it will be 5 if it is 4 then it'll be 4 because after every iteration
# we are getting 1 value fixed.
            if nums[j] > nums[j+1]: # Ascending Order
                # Swap
                temp = nums[j]
                nums[j] = nums[j+1]
                nums[j+1] = temp
                # Internal loop does the swapping and the outer loop is just to check the number of elements you have - number of iterations depend upon it. - for 6 values - 5 iterations so after every iteration we have a value fixed at the end.
nums = [5,3,8,6,7,2]
# We can use inbuilt sorting techniques but we are doing manually here.
sort(nums)
print(nums)
```

```python
# Selection Sort: What's wrong with bubble sort the problem is in every iteration first of all we do multiple iterations right 
# because we have to make sure that you get sorted values and at one point you swap two values the thing is in each iteration
# you do multiple swapping right if you remember we have done multiple swapping so there's nothing wrong in traversing from 
# start to end but swapping consumes a lot of processing power so we don't want to do that we don't want to consume our CPU 
# power and memory to do that so swapping should be done only once and that's where selection sort comes into picture so in 
# selection sort what we do is we go from start to end and we find either min value or max - depend upon different 
# implementation in this example we'll go for the min value so what you will do you will go from start to end and you will find
# the min value so in each iteration you will do only one thing which is you will find the mean value. 

# At the end of the first iteration, we get the min value which is 2 in [5,3,8,6,7,2] - now we swap 2 with 5 as we started with
# 5 and now after the first iteration we know 2 is the min value at the start and it is fixed - so we are getting 2 different
# arrays here - one is sorted and the other is unsorted and 2 is sorted one and other remaining ele is unsorted one and in the
# remaining ele we do the same thing find the min value and swap. - in each iteration we go from start to end and we do swapping
# only once and with each iteration we start skipping the values - the sorted ones - after i'th iteration we skip i values.

# Code:

def sort(nums):
    for i in range(5): # higher index -1 = = 6-1 = 5 = len(nums) -1 as we use i+1 in inner loop imo # What range we can define - we can go for min value or max value when we go with max value we go from the highest index to the lowest index. In this case we find min value so we go from lower index to the higher index.
        minpos = i
# so normally what we do is we say mean position is 0 right but we're not doing in this case because after every iteration we 
# are fixing the sorted array right and in that the zeroth index is gone so we cannot simply reuse it and that's why I'm 
# changing the min position to i everytime after every iteration 
        for j in range(i+1,6): # so after every iteration you can see we are creating that sorted array right so we have to reduce the
# size of our unsorted array and that's why the range of this will be lowered so we'll say range it would start from i right and
# you have to go till 6 because the size is 6 it will go up to 5 the inner loop okay.
            # Only thing to do here is to find the min value.
            if nums[j] < nums[minpos]:
                minpos = j
        
        temp = nums[i]
        nums[i] = nums[minpos]
        nums[minpos] = temp
        # Swap after getting the min pos. We are taking a temporary variable and then we are swapping the index value I because
# if you'll see we are going from start to end and we have to make sure that after every iteration you are getting one fixed 
# value so you will not simply keep swapping the elements every time in the j loop(like in bubble sort) you will do that in the
# outer loop so you will swap only once and that's why we are taking i index and not j so you're swapping the index i with the
# min position and at the end you will get the sorted array.
        print(nums) # See after every iteration how it looks like.
nums = [5,3,8,6,7,2]
sort(nums)
print(nums)
```

```python
# Zip function: 

# when you say zip what it means exactly the zip you have in your jeans right so normally it connects multiple components so 
# let's say if you have two list and if you want to join them like this (parallely) you can use zip:

# 2 list of 2 tuples.

names = ("Navin","Kiran","Harsh")
comps = ("Dell","Apple", "MS") # Respectively

zipped = zip(names,comps)
print(zipped)
```

```python
# We are getting the type as zip.
zipped = list(zip(names,comps))
print(zipped)
```

```python
# We can use set also only diff is set will be unstructured/unordered and it'll only give unique values.
zipped = set(zip(names,comps))
print(zipped)
```

```python
names = ("Navin","Kiran","Harsh")
comps = ("Dell","Apple", "MS") 
zipped = dict(zip(names,comps)) # Dictionary format
print(zipped)
```

```python
# Can also use an iterator:  or a loop too and we can iterate between the values.
names = ("Navin","Kiran","Harsh")
comps = ("Dell","Apple", "MS") 
zipped = zip(names,comps)
for (a,b) in zipped:
    print(a,b)
```
